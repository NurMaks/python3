```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('mtcars.csv')
data
```

```python
y_pos = np.arange(len(data["model"][:15]))
plt.figure(figsize=(25,7))
plt.bar(y_pos, data['mpg'][:15])
plt.xticks(y_pos, data['model'][:15])
plt.xlabel("Model", fontsize=20)
plt.ylabel("MPG", fontsize=20)
plt.title("Fuel consumption of certain cars", fontsize=25)
plt.show()
```

```python
plt.boxplot(data["disp"])
plt.xlabel("Displacement", fontsize=12)
plt.show()
```

```python
df = pd.read_csv("wb.csv")
df = df.drop(columns=["Time", "Time Code"])
df.columns = ["CN", "CC", "GDP", "PT", "LE"]
df["GDP"].replace("..", np.nan, inplace=True)
df["PT"].replace("..", np.nan, inplace=True)
df["LE"].replace("..", np.nan, inplace=True)
df.dropna(inplace=True)
df = df[:5]
df
```

```python
plt.pie(df["PT"], labels=df["CC"])
my_circle=plt.Circle( (0,0), 0.6, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title("Population of some countries", fontsize=13)
plt.show()
```
